#### Data:

The data used for this homework was taken from The Network Data Repository. The title of these data was "aves-geese-female-foraging". This network dataset recorded animals grazing on the same patch during a sampling period (15 days). They were assumed to be associating (gambit of the group). 

```python
import networkx as net

o=net.Graph()

G_edges = net.read_edgelist("aves-geese-female-foraging.edges", nodetype=int, data=(("Type", str),))

```

The number of nodes in this dataset are 20. The number of animals observed in this dataset is 20. 

```python
G_edges.number_of_nodes()
```

The number of edges in this graph are 190. There are 190 connections in this network. 

```python
G_edges.number_of_edges()
```

The diameter of the graph is 1. 

net.diameter(G_edges)


```python
net.draw(G_edges, with_labels=True, font_weight='bold')
```
